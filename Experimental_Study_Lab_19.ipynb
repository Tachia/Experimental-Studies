## **INTRODUCTION**
### **Definitions: Electron-Beam Plasma (EBP):**

Electron beam plasma is a type of plasma that is generated by directing an electron beam at a gas or solid target. The high-energy electrons in the beam ionize the target atoms or molecules, creating a plasma. Electron beam plasma is commonly used in a variety of applications, including surface modification, materials processing, and plasma-based devices. The properties of electron beam plasma, such as its density, temperature, and composition, can be controlled by adjusting the parameters of the electron beam, such as its energy and current.

Electron Beam Plasma (EBP) is a high-energy plasma produced by bombarding a target material with a focused beam of high-energy electrons. The high-energy electrons collide with the target atoms and cause ionization, leading to the formation of plasma. The plasma is highly ionized and has a high electron temperature, which makes it an attractive medium for various applications such as material processing, surface treatment, and plasma-based diagnostics.

EBP can be produced using different electron beam sources, including linear accelerators, electron guns, and microwave sources. The parameters of EBP, such as plasma density, electron temperature, and electron energy, can be controlled by adjusting the electron beam parameters and the operating conditions.

Conclusion: EBP has been used in various applications such as thin film deposition, etching, and surface modification. It is also used in the production of nanoparticles, the study of plasma-surface interactions, and the analysis of complex materials.
In summary, Electron Beam Plasma is a high-energy plasma produced by bombarding a target material with a focused beam of high-energy electrons, which has a wide range of applications in material processing, surface treatment, and plasma-based diagnostics.


### **Basic properties of the Electron-Beam Plasma:**

The properties of electron beam plasma depend on various factors such as the electron energy, the pressure and composition of the background gas, and the presence of any magnetic or electric fields. Some common properties of electron beam plasma include:
* High Temperature: Electron beam plasmas are often very hot, with temperatures ranging from a few thousand to tens of thousands of Kelvin.
* High Electron Density: The electron beam ionizes the background gas, creating a region of high electron density.
* Non-Thermal Distribution: The electron energy distribution in an electron beam plasma is often non-thermal, meaning that the electrons have a wide range of energies, rather than a single temperature.
* Strong Electric Fields: Electron beam plasmas can generate strong electric fields due to the large difference in potential between the electrons and the background gas.
* Hydrodynamic Behavior: Electron beam plasmas can exhibit hydrodynamic behavior, where the plasma responds to changes in pressure and temperature, leading to fluid-like behavior.
* Particle Acceleration: Electron beam plasmas can be used for particle acceleration, as the high energy electrons can transfer energy to other particles, leading to acceleration.

These properties can lead to a range of interesting phenomena in electron beam plasmas, such as the formation of filaments and instabilities, and the generation of intense light and radiation.


### **Electron Beam (EB) Propagation in Dense Media:**

Fast electrons range in dense media refers to the maximum distance that fast electrons can travel in a dense material before losing their energy. In general, fast electrons are defined as electrons with kinetic energy greater than a few electron volts (eV). When fast electrons collide with the atoms and molecules in a dense material, they transfer their energy to the surrounding media, which leads to the creation of secondary electrons and the ionization of the medium.

The fast electron range in dense media is influenced by several factors, including the density of the material, the type of material, the energy of the fast electrons, and the presence of impurities. In general, the fast electron range in dense media decreases as the density of the material increases and as the energy of the fast electrons decreases. The fast electron range is also influenced by the presence of impurities, as impurities can absorb fast electrons and cause them to lose their energy more quickly.

In some applications, such as ion implantation, fast electron range in dense media is a critical parameter, as it determines the depth to which the ions will penetrate into the material. In other applications, such as radiation therapy, the fast electron range is an important parameter, as it determines the maximum depth to which the therapeutic dose of radiation will penetrate into the body.


### **Bethe-Bloch Formula:**

The Bethe-Bloch formula is a fundamental equation in the field of nuclear physics that is used to describe the energy loss of high-energy charged particles as they traverse a dense material. It is particularly useful for describing the energy loss of fast electrons in dense media, including the deceleration of these electrons.

**The Bethe-Bloch formula is given by:**

For a given particle with speed v, charge z, (in multiples of the electron charge), and energy E, travelling a distance x into a target of electron number density n, and mean excitation energy I, the relativistic version of the formula is given by:

```python
from IPython.display import display, Math

formula = r"-\langle\frac{dE}{dx}\rangle = \frac{4\pi}{m_e c^2}\frac{nz^2}{\beta^2}" \
          r"\left(\frac{e^2}{4\pi\epsilon_0}\right)^2" \
          r"\left[\ln\left(\frac{2m_e c^2\beta^2}{I(1-\beta^2)}\right)-\beta^2\right] ........... .....  [1]"

display(Math(formula))
```

where **dE/dx** is the energy loss per unit path length, **K** is a constant that depends on the properties of the material, **Z** is the atomic number of the material, **β** is the velocity of the particle divided by the speed of light, **v** is the velocity of the particle, **m** is the mass of the particle, and **I** is the mean excitation energy of the material.

***Note that:***


```python
from IPython.display import display, Math

formula = r"β = \frac{V}{c} ................ [2]"

display(Math(formula))
```



The equation shows that the energy loss of fast electrons in dense media is proportional to the square of the atomic number of the material, the inverse of the particle velocity squared, and a logarithmic term that depends on the mean excitation energy of the material.

The term (1/β^2)*(1/v^2) represents the slowing down of the particle, which results in the deceleration of the fast electron.

The electron density of the material can be calculated by:

```python
from IPython.display import display, Math

formula = r"n = \frac{N_A*Z*ρ}{A*M_u} ........... .....  [3]"

display(Math(formula))
```

where **ρ** is the density of the material, **Z** is the atomic number, **A** is the relative atomic mass, **N_A** is the Avogadro number and **Mu** is the molar mass constant.

For low energies, that is for small velocities of the particles **β << 1**, the Bethe formula reduces to:


```python
from IPython.display import display, Math

formula = r"-\langle\frac{dE}{dx}\rangle = \frac{4\pi nz^2}{m_e V^2}" \
          r"\left(\frac{e^2}{4\pi\epsilon_0}\right)^2" \
          r"\left[\ln\left(\frac{2m_e V^2}{I}\right)\right] ........... ..... ...[4]"

display(Math(formula))
```

This can be seen by first replacing **βc** with **V** in equation [2] and then neglecting **β^2** because of its size.

In conclusion, the Bethe-Bloch formula is a useful tool for describing the energy loss and deceleration of fast electrons in dense media. The equation can be used to predict the behavior of fast electrons in different materials, and to understand the factors that influence the deceleration of these electrons.

### **Discussion.**

**Physical explanation of experimental data and B (Pm, z)-functions (z – EB injection axis)**

The experimental data collected during the study showed that the brightness of the plasma increased as the gas pressure decreased. This result can be explained by the fact that as the pressure decreases, the mean free path of the gas molecules increases, leading to fewer collisions between the gas molecules. This results in a decrease in the number of neutral particles in the plasma, allowing the electrons to travel a longer distance without losing energy through collisions with neutral particles. This leads to an increase in the electron density, which results in an increase in the plasma brightness. The B(P) function that was recorded for different sections of the quartz tube showed that the plasma brightness increased as the gas pressure decreased. This result was consistent with the brightness-pressure plot obtained from the photodiode sensor. Additionally, the B(z) function constructed using specially developed software showed that the brightness of the plasma varied along the length of the tube, with the brightest region of the plasma being close to the electron beam injection axis. 

This observation can be explained by the fact that the electron density in the plasma is highest near the electron beam injection axis. The electrons that are injected into the plasma by the electron beam generate an ionization wave that propagates away from the injection axis, creating the plasma. The plasma density and brightness are highest near the injection axis due to the high electron density, which results in a higher ionization rate and a higher plasma density. As the ionization wave propagates away from the injection axis, the plasma density and brightness decrease due to a decrease in the electron density and a lower ionization rate. 

In summary, the experimental data and the B (Pm, z)-functions obtained in this study provide valuable insights into the behavior of electron beam plasma (EBP) generated in a glass quartz tube. The results suggest that the plasma brightness is influenced by the gas pressure, and the plasma density and brightness are highest near the electron beam injection axis. These findings have important implications for the design and optimization of plasma-based devices and applications.
