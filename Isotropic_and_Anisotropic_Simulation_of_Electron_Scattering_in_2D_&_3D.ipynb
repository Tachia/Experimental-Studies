MONTE CARLO SIMULATION TASK

QUESTION:

Show the trajectory of an electron motion in 2D using Monte Carlo Simulation, and hence determine the angle, phi (φ) and the mean free path, lambda (Λ) for: ten (10), hundred (100) and thousands (1000, 10000) collisions.

Given that:

φ = 2πa, a --- [0,1] "random valuable"

Λ = -λ.ln(a)

NB: The value of "a" must be selected differently for each collision(s).

where: Phi, φ - is the angle between the electron collisions with other particles, a - is Bohr radius, lambda, Λ - is the mean free path, λ - is lambda null which is assumed to be 1, ln(a) - is the natural logarithsm of a in base e.

Simulation for ten (10) Collisions

```python
#@title
import numpy as np
import matplotlib.pyplot as plt
import random

numbere = 1
num_a = 10
e = []
for j in range(numbere):

  a = np.random.uniform(size=num_a)
  b = np.random.uniform(size=num_a)
  ex = [0]
  ey = [0]
  for i in range(len(a)):
    Lambda = -np.log(b[i])
    phi = 2*np.pi*a[i]

    ex.append(Lambda*np.cos(phi))
    ey.append(Lambda*np.sin(phi))
  plt.plot(ex, ey, linestyle='-',color ="navy", linewidth = 3, marker='o', markerfacecolor='green', markersize=10)
plt.show()

# Graph of the simulation
L = []
a = np.random.uniform(size=num_a)
b = np.random.uniform(size=num_a)
ex = [0]
ey = [0]

for i in range(len(a)):
  Lambda = -np.log(b[i])
  phi = 2*np.pi*a[i]

  ex.append(Lambda*np.cos(phi))
  ey.append(Lambda*np.sin(phi))
lx = 0
ly = 0
N = []
n = 1
for i in range(len(a)):
  N.append(n)
  n = n + 1
  lx = lx + ex[i]
  ly = ly + ey[i]
  L.append((lx**2+ly**2)**(1/2))

plt.plot(N, L, color ="darkblue")
plt.title("The Plot of Distance from the Source against Number of Collisions")
plt.xlabel("Number of Collisions")
plt.ylabel("Distance from the Source")
plt.show()

# Absolute value of the graph
L = []
a = np.random.uniform(size=num_a)
b = np.random.uniform(size=num_a)
ex = [0]
ey = [0]

for i in range(len(a)):
  Lambda = -np.log(b[i])
  phi = 2*np.pi*a[i]

  ex.append(Lambda*np.cos(phi))
  ey.append(Lambda*np.sin(phi))
lx = 0
ly = 0
N = []
n = 1
for i in range(len(a)):
  N.append(n)
  n = n + 1
  lx = lx + abs(ex[i])
  ly = ly + abs(ey[i])
  L.append((lx+ly)**(1/2))

plt.plot(N, L, color ="darkblue")
plt.title("The Plot of Distance from the Source against Number of Collisions: Absolute Values")
plt.xlabel("Number of Collisions")
plt.ylabel("Distance from the Source")
plt.show()
```

```python
#@title
import numpy as np
from scipy import stats
import numpy
import pandas as pd

numpy.seterr(divide = 'ignore')
y = np.log(L)
x = np.log(N)
plt.plot(x,y, c = "navy")
plt.title("The graph of log(L) versus log(N)", fontsize = "20")
plt.xlabel("log(N)", fontsize = "20")
plt.ylabel("log(L)", fontsize = "20")
plt.grid()

print("The min. and max. values of log(L) and log(N):",(min(x), max(x), min(y), max(y)))
print("\n")
p = max(y)
l = (max(x) - min(x))
n = (max(y) - min(y))
m = (p/l)
print("The Slope of log(L) versus log(N):", m)
print("\n")
```

QUESTION:

Show the trajectory of an electron motion in 3D using Monte Carlo Simulation, and hence determine the angles, phi (φ), theta (θ) and the mean free path, lambda (Λ) for: ten (10), hundred (100) a thousand (1000) and tens of thousand (10000) collisions.

Given that:

Phi (φ) = 2πa, a --- [0,1] "random valuable"

Theta (θ) = arccos(1-2a)

Lambda (Λ) = -λ.ln(a)

NB: The value of "a" must be selected differently for each collision(s).

where: Phi, φ and theta θ, are the angles between the electron collisions with other particles, a - is Bohr radius, lambda, Λ - is the mean free path, λ - is lambda null which is assumed to be 1, ln(a) - is the natural logarithsm of a in base e.

```python
#@title
import numpy as np
import matplotlib.pyplot as plt
import random

numbere = 1
num_a = 10
e = []
for j in range(numbere):
  ax = plt.axes(projection="3d")
  a = np.random.uniform(size=num_a)
  b = np.random.uniform(size=num_a)
  c = np.random.uniform(size=num_a)
  ex = [0]
  ey = [0]
  ez = [0]
  for i in range(len(a)):
    Lambda = -np.log(b[i])
    phi = 2*np.pi*a[i]
    theta = np.arccos(1 - 2*c[i])

    ex.append(Lambda*np.cos(phi))
    ey.append(Lambda*np.sin(phi))
    ez.append(Lambda*theta)
  plt.plot(ex, ey, ez, linestyle='-',color ="navy", linewidth = 3, marker='o', markerfacecolor='green', markersize=10)
plt.show()

# Absolute value of the graph
L = []
a = np.random.uniform(size=num_a)
b = np.random.uniform(size=num_a)
c = np.random.uniform(size=num_a)
ex = [0]
ey = [0]
ez = [0]

for i in range(len(a)):
  Lambda = -np.log(b[i])
  phi = 2*np.pi*a[i]
  theta = np.arccos(1 - 2*c[i])

  ex.append(Lambda*np.cos(phi))
  ey.append(Lambda*np.sin(phi))
  ez.append(Lambda*theta)
lx = 0
ly = 0
lz = 0
N = []
n = 1
for i in range(len(a)):
  N.append(n)
  n = n + 1
  lx = lx + abs(ex[i])
  ly = ly + abs(ey[i])
  lz = lz + abs(ez[i])
  L.append(((lx+ly+lz)**(1/2)))

plt.plot(N, L, color ="darkblue")
plt.title("The Distance from the Source against Number of Collisions: Absolute Values")
plt.xlabel("Number of Collisions")
plt.ylabel("The Distance from the Source")
plt.show()
```

MONTE CARLO SIMULATION (3D Simulation)

Point source is anisotropic: all electrons are emitted along one direction; electron-atom scattering is anisotropic: Theta, θ is determined from sin(θ/2) = a_3, (where a_1, a_2 and a_3 are random numbers selected between 0 and 1)

Determination of λ and φ is given by: Lambda, λ = -lna_1, phi, φ = 2πa_2.

It is necessary to make simulation for the anisotropic case and to compare results with previous results for isotropic case. Lambda is the length between the collisions, phi and theta are the angular displacement in x-y, x-z coordinates respectively.

From the sine identity formula, we have that:

sin(θ/2) = a_3

θ = 2arcsin(a_3)

```python
#@title
import numpy as np
import matplotlib.pyplot as plt
import random

numbere = 1
num_a = 100
e = []
for j in range(numbere):
  ax = plt.axes(projection="3d")
  a = np.random.uniform(size=num_a)
  b = np.random.uniform(size=num_a)
  c = np.random.uniform(size=num_a)
  a = c
  ex = [0]
  ey = [0]
  ez = [0]
  for i in range(len(a)):
    Lambda = -np.log(a[i])
    phi = 2*np.pi*b[i]
    theta = np.pi*c[i]

    X = (Lambda*np.sin(theta)*np.cos(phi)) 
    Y = (Lambda*np.sin(theta)*np.sin(phi)) 
    Z = (Lambda*np.cos(theta)) 

    ex.append(X)
    ey.append(Y)
    ez.append(Z)
  plt.plot(ex, ey, ez, linestyle='-',color ="navy", linewidth = 3, marker='o', markerfacecolor='green', markersize=10)
plt.show()

L = []
lx = 0
ly = 0
lz = 0
N = []
n = 1
for i in range(len(a)):
  N.append(n)
  n = n + 1
  lx = lx + (ex[i])
  ly = ly + (ey[i])
  lz = lz + (ez[i])
  L.append(((lx+ly+lz)**(1/2)))

plt.plot(N, L, color ="darkblue")
plt.title("The Distance from the Source against Number of Collisions: Absolute Values")
plt.xlabel("Number of Collisions")
plt.ylabel("The Distance from the Source")
plt.show()
```

```python
#@title
import numpy as np
from scipy import stats
import numpy
import pandas as pd

numpy.seterr(divide = 'ignore')
y = np.log(L)
x = np.log(N)
plt.plot(x,y, c = "navy")
plt.title("The graph of log(L) versus log(N)", fontsize = "20")
plt.xlabel("log(N)", fontsize = "20")
plt.ylabel("log(L)", fontsize = "20")
plt.grid()

print("The min. and max. values of log(L) and log(N):",(min(x), max(x), min(y), max(y)))
print("\n")
p = max(y)
l = (max(x) - min(x))
n = (max(y) - min(y))
m = (p/l)
print("The Slope of log(L) versus log(N):", m)
print("\n")
```
