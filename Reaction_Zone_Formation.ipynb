**ABSTRACT.**

An experimental setup for generating electron-beam and hybrid plasmas of gases and vapor gas mixtures with particles of dispersed powders and liquid droplets additives has been created and fully tested. The setup is a prototype of a plasma-chemical reactor for "green" technologies based on cold non-equilibrium plasmas.


**Keywords: **

Electron beam plasma, hybrid plasma, plasma-chemical reactor of fore vacuum pressure, "green" technologies.

**INTRODUCTION**

Beam-plasma reactors, in which the reaction volume is formed by electron beams injected into a fairly dense gaseous environment, have occupied a well-defined niche among the plasma systems used for the production of materials, surface treatment of products, processing of natural organic raw materials, purification and conversion of various gases [1, 2]. The competitive advantages of such reactors are based on the essential features of the principle of electron beam plasma generation, namely: 
* The electron beam (EB) can be injected into gas and vapor-gas mixtures of almost any component composition, and the pressure of the plasma-forming medium (P_m) can vary over a wide range from the lower boundaries of the fore vacuum range (~10-3 Torr) to atmospheric. 
* In the fore vacuum pressure range, electron beam plasma (EBP) has a high chemical activity even at low, up to room temperatures. High rates of plasma-stimulated processes can significantly reduce the residence time of the substance in the reaction volume, minimize the thermal effect of EB on the substance and thereby prevent thermal damage to materials, for example, thermolabile polymers. 
* The objects of targeted beam-plasma exposure can be almost any substances in all aggregate states: gases, vapors, liquids, macroscopic solids and finely dispersed aerosols containing powder particles or liquid droplets. 
* EB injection is compatible with the action of other ionizers, for example, gas discharges of different frequency ranges. In this case, it is possible to generate the so-called hybrid plasma (HP).  

The variety of conditions for the generation of EBPs and EB, as well as the design and technological features of beam-plasma reactors, determines the range of physical and engineering tasks that need to be solved when designing such devices and optimizing their operating modes. The formation of a stable and well-controlled reaction volume is one of the complex problems, on the success of the solution of which the technical and economic characteristics of the equipment largely depend. 


**RESEARCH METHODS** 

In beam-plasma reactors, one of several variants of the reaction volume is usually implemented, differing in the phase composition of the plasma-forming medium and the scheme of organization of the working process. The greatest difficulties arise in cases where:  
* The plasma-forming medium is an aerosol containing a liquid or solid dispersed phase; 
* EBP is in contact with a dielectric, for example, with the surface of the treated substance (in solid or liquid phase), the walls of the reaction chamber or elements of its internal equipment; 
* It is necessary to ensure the joint operation of several ionizers, for example, to combine the EB injector with the gas discharge system. 
* In the reaction zone, an EBP or EB flow is required. 

One of the variants of the laboratory sample of the reactor, on which it is possible to work out real beam-plasma technologies, is illustrated by Fig. 1. The sharply focused EB 1 is formed in the high-vacuum chamber 3 by an electron beam gun (not shown in the figure) and directed to the input of the output device 2, through which it is injected into the cylindrical reaction chamber 7 along its axis. The output device is structurally combined with a nozzle 4 forming a gas stream 6 partially or completely filling the reaction chamber. Plasma-forming gases with specified flows G1 and G2 are fed into the nozzle pre-chamber, where they are mixed.  

The opposite end of the reaction chamber is connected to a forevacuum pump 12. A PID pumping capacity regulator G0 (9)  is installed in the vacuum line. The flow rate w and the pressure value in the reaction chamber Pm can be controlled by selecting the values G1, G2 and G0.

The gas flow in the reaction chamber is ionized and excited by the electron beam, i.e. plasma is generated, the properties of which depend on the parameters of the injected EB, the characteristics of the plasma-forming mixture and the geometry of the reaction chamber. If necessary, a powder or liquid may be introduced into the plasma stream using the nebulizer 5. In this case, the flow of the plasma-forming medium becomes heterophase. 

Gas-phase and heterophane plasma chemical reactions lead to the formation of target and by-products, the concentrations of which must be measured. For this purpose, a mass spectrometer 11 equipped with a refrigerator and a throttle device 10 is installed on the pumping line.  
If required, plasma stream generation can also be carried out in free space. To do this, it is sufficient to remove the reaction chamber 7, replacing it with a working vacuum chamber of large volume 8. Pumping of gas flowing into the working chamber is carried out by an independent pump 13 through a PID controller. Such a design version of the reactor turned out to be the most convenient for experiments with macroscopic bodies, which can be placed both in a free plasma stream and in a practically stationary cloud of EBP.  In the latter case, it is enough to minimize the costs of G1, G2 and G0. Unfortunately, in a circuit solution with a large-volume reaction chamber, mass spectroscopy of the reaction volume becomes a much more difficult task: the gaseous sample must be taken directly from the reaction zone using special probes and a special differential pumping system. 

To generate an EB, an electrode system (not shown in the figure) is installed outside or inside the cylindrical working chamber 7, which creates an additional electromagnetic field in the reaction volume. In our experiments, the electrode system was connected to a RF generator, i.e. the EB was a superposition of electron beam plasma and high-frequency gas discharge plasma of the capacitive type. Similarly, hybrid plasma can be generated in free space (see above).

**DISCUSSION**

Let us consider several practical problems that have been successfully solved on beam-plasma installations with a heterophase working volume operating in the fore vacuum pressure range. 

1)	Control of the dynamics of finely dispersed aerosol particles in the reaction volume. 
 
The Coulomb interaction of fine particles in beam-plasma systems under deep vacuum conditions and at forvalacum pressures must be taken into account in the design and operation of the most diverse devices based on the use of EBP and EB. These phenomena have been studied in sufficient detail in relation to the electrostatic charging of spacecraft under the influence of the solar wind, as well as during the operation of electric rocket engines and on-board power plants [3, 4]. For terrestrial conditions, the electrostatic charging of micron and submicron-sized particles was studied mainly in connection with the design of technological equipment, for example, plasma-chemical reactors for various purposes.  

When dispersed powders or liquid droplets are the active medium in a beam-plasma reactor, the Coulomb interaction of these particles with each other and with the structural elements of the reaction chamber gives rise to a complex complex of electrophysical processes and phenomena of energy and mass transfer. One such phenomenon is the intense removal of condensed dispersed phase particles to the periphery of the reaction volume (see Figure 3a), which can disrupt the operation of the plasma-chemical reactor. This effect is observed at sufficiently low pressures in the reaction chamber (Pm < 0.1 Torr), and an increase in pressure to several Torr is able to suppress the electrostatic dispersion of the particles of the dispersed phase (see Figure 3b). It is important that as P m increases, the transition from an unstable mode of operation of the reactor to a stable one occurs smoothly, and the boundaries of the transient pressure range depend on the chemical composition of the plasma-forming gas and the constructive features of the reaction chamber. 

Opposite effects are possible and have been repeatedly observed in experiments, for example, the self-organization of plasma-dust structures in hybrid plasma [5]. The effect of EP injection into the plasma-dust structure is an independent research task directly related to the problem of controlling the heterophase plasma volume, in which the condensed dispersed phase is distributed in an orderly manner. Such a problem was solved in relation to a hybrid reactor with a heterophase plasma-forming medium in the form of stable plasma-dust structures levitating above the surface of a secluded flat RF electrode. The procedure for creating ordered plasma-dust structures is as follows. The material from which the plasma-dust structure (for example, the sublimating structure) will be formed in the future was evaporated by an EP pulse. The vapor activated by the beam gradually condensed in the form of finely dispersed clusters, which were captured by the electromagnetic field of the RF electrode. At the end of the steam condensation process, a stable dust structure was formed above the surface of the electrode, which can be controlled by spatial scanning of the EP in the RF discharge zone or by pulse-periodic exposure to the beam. The characteristic pressure range in the reaction volume, when such control methods are effective, is 0.5-2.0 Torr. 

2) Formation of gas-liquid flows of EBP and EB. 
In the fore vacuum pressure range, the velocity pressure of the plasma flow is sufficient for pneumatic spraying of the liquid injected into it in the form of a stream or a thin veil. Spraying is possible both in a free jet of gas excited by an electron beam (see Figure 4a) and inside the channel in which the plasma stream is formed. In the latter case, it is possible to obtain a stream of EB of various gases containing vapors and droplets of liquid. For this purpose, active and passive elements of a special electrode system are installed outside and inside the cylindrical reaction chamber; THE active electrodes of this system are supplied with RF power. As an example, Figure 4b shows a photograph of the gas-liquid flow of a EBU whose components are methane (carrier gas) and partially evaporated methyl alcohol. Characteristic conditions of experiments of this series: Pm ~ 5 Torr, w ~ 10 m / s. 
 
Forevacuum beam-plasma reactors of the hybrid type with a gas-liquid reaction volume are currently being tested in order to identify the possibilities of their use in promising "green" technologies based on plasma-stimulated conversion of hydrocarbons, gas-liquid conversion (GTL-processes) in plasma jets, reforming of renewable organic raw materials and natural biopolymers.
